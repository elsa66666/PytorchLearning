```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('dataset/credit-a.csv', header=None)
```

```python
data
```

```python
x_values = data.iloc[:, :-1] #取所有行所有列
y_values = data.iloc[:, -1].replace(-1,0)  #取所有行，最后一列
```

```python
X = torch.from_numpy(x_values.values).type(torch.float32)
Y = torch.from_numpy(y_values.values.reshape(-1, 1)).type(torch.float32)
```

```python
from torch import nn
```

```python
#创建模型
model = nn.Sequential(
        nn.Linear(15,1), #输入15个属性，输出1个属性
        nn.Sigmoid() #激活层，使所有层顺序连接
)
```

```python
model
```

```python
loss_fn = nn.BCELoss() #二元交叉熵损失
opt = torch.optim.Adam(model.parameters(), lr=0.001)
batches = 16  #16个数据为一组
number_of_batch = 653//16 #全部数据训练一次需要多少组
epoches = 1000 #循环1000次

for epoch in range(epoches):
    for i in range(number_of_batch):
        start = i*batches
        end = start + batches
        x = X[start:end]
        y = Y[start:end]
        y_pred = model(x)
        loss = loss_fn(y_pred, y)
        opt.zero_grad()
        loss.backward()
        opt.step()
```
